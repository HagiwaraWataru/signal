```python
%matplotlib nbagg

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import math
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('signal_csv/''wk''.csv')
```

```python
df
```

```python
df = df.rename(columns={"Unnamed: 0":"time"})
df
```

```python
time = df.time
r1 = df.rever1.values
r2 = df.rever2.values
r3 = df.rever3.values
```

```python
fig = plt.figure(figsize=(4,4), facecolor='w')
ax = fig.add_subplot(111)
ax.plot(time, r1)
ax.plot(time, r2)
ax.plot(time, r3)
plt.show()
```

```python
fig = plt.figure()
l = 5000

ims = []
for i in range(100):
    x = l*math.sin(r1[i*10]/180*math.pi)
    y = l*math.cos(r1[i*10]/180*math.pi)

    im = plt.plot([0,x],[0,y])
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims)
plt.show()
```
